As variáveis `ANO` e `CO_MUNICIPIO` aparecem na análise estatística por serem numéricas, mas não representam atributos explicativos. O ano é apenas a referência temporal do estudo e o código do município é utilizado como identificador para integração das bases.

## Análise Regional: Região Geográfica Intermediária de Uberaba

Nesta etapa, será analisada a Região Geográfica Intermediária de Uberaba, recorte territorial no qual está inserido o município de Frutal.

A escolha desse recorte regional se justifica por ser uma divisão territorial oficial, por incluir Frutal e por permitir uma comparação mais específica do município com uma região próxima, sem limitar a análise apenas aos municípios imediatamente vizinhos.

A análise seguirá a lógica do geral para o específico: primeiro os municípios de Minas Gerais, depois a Região Geográfica Intermediária de Uberaba e, posteriormente, o município de Frutal.

```python
import unicodedata

def remover_acentos(texto):
    texto = str(texto)
    texto = unicodedata.normalize("NFKD", texto)
    texto = texto.encode("ASCII", "ignore").decode("utf-8")
    return texto.upper().strip()

municipios_rgint_uberaba = [
    "Água Comprida",
    "Araxá",
    "Campo Florido",
    "Campos Altos",
    "Carneirinho",
    "Comendador Gomes",
    "Conceição das Alagoas",
    "Conquista",
    "Delta",
    "Fronteira",
    "Frutal",
    "Ibiá",
    "Itapagipe",
    "Iturama",
    "Limeira do Oeste",
    "Nova Ponte",
    "Pedrinópolis",
    "Perdizes",
    "Pirajuba",
    "Planura",
    "Pratinha",
    "Sacramento",
    "Santa Juliana",
    "Santa Rosa da Serra",
    "São Francisco de Sales",
    "Tapira",
    "Uberaba",
    "União de Minas",
    "Veríssimo"
]

municipios_rgint_normalizados = [
    remover_acentos(municipio)
    for municipio in municipios_rgint_uberaba
]

df["NO_MUNICIPIO_NORMALIZADO"] = df["NO_MUNICIPIO"].apply(remover_acentos)

df_rgint_uberaba = df[
    df["NO_MUNICIPIO_NORMALIZADO"].isin(municipios_rgint_normalizados)
].copy()

print("Municípios esperados:", len(municipios_rgint_uberaba))
print("Municípios encontrados no dataset:", df_rgint_uberaba["NO_MUNICIPIO"].nunique())

df_rgint_uberaba[
    [
        "NO_MUNICIPIO",
        "MEDIA_INSE",
        "TDI_ANOS_FINAIS",
        "MEDIA_ALUNOS_TURMA_ANOS_FINAIS",
        "TAXA_ABANDONO_ANOS_FINAIS"
    ]
].sort_values("NO_MUNICIPIO").reset_index(drop=True)
```

```python
variaveis_regionais = [
    "MEDIA_INSE",
    "TDI_ANOS_FINAIS",
    "MEDIA_ALUNOS_TURMA_ANOS_FINAIS",
    "TAXA_ABANDONO_ANOS_FINAIS"
]

comparacao_regional = pd.DataFrame({
    "Indicador": variaveis_regionais,
    "Media MG": df[variaveis_regionais].mean().values,
    "Mediana MG": df[variaveis_regionais].median().values,
    "Media RG": df_rgint_uberaba[variaveis_regionais].mean().values,
    "Mediana RG": df_rgint_uberaba[variaveis_regionais].median().values
})

comparacao_regional["Diferença para Média MG"] = (
    comparacao_regional["Media RG"] - comparacao_regional["Media MG"]
)

comparacao_regional
```

### Interpretação da comparação regional

A comparação entre Minas Gerais e a Região Geográfica Intermediária de Uberaba mostra que a região apresenta `MEDIA_INSE` superior à média estadual. Enquanto a média de Minas Gerais é de 4,9382, a média regional é de 5,1873. Isso indica que, em termos socioeconômicos, os municípios da região analisada apresentam, em média, condição mais favorável do que o conjunto dos municípios mineiros.

Por outro lado, a Taxa de Distorção Idade-Série dos Anos Finais é consideravelmente mais elevada na Região Intermediária de Uberaba. A média estadual é de 11,86%, enquanto a média regional chega a 17,36%, uma diferença de aproximadamente 5,5 pontos percentuais. Esse resultado indica que, apesar do INSE mais alto, a região apresenta maior proporção de estudantes em atraso escolar nos Anos Finais.

A Média de Alunos por Turma também é ligeiramente superior na região. Minas Gerais apresenta média de 24,49 alunos por turma, enquanto a Região Intermediária de Uberaba possui média de 25,77. Essa diferença não é muito elevada, mas indica que as turmas da região tendem a ser um pouco mais cheias do que a média estadual.

A Taxa de Abandono Escolar dos Anos Finais apresenta média de 1,62% na Região Intermediária de Uberaba, valor um pouco acima da média estadual de 1,47%. A mediana, entretanto, é igual nos dois recortes, com valor de 1,1%. Isso indica que o abandono escolar na região é próximo ao padrão estadual, embora a média regional seja levemente superior.

De modo geral, a Região Geográfica Intermediária de Uberaba apresenta um perfil interessante: possui nível socioeconômico médio mais alto que Minas Gerais, mas também apresenta Taxa de Distorção Idade-Série mais elevada. Esse resultado reforça a importância de analisar o abandono escolar considerando múltiplos indicadores, pois condições socioeconômicas mais favoráveis não eliminam, por si só, desafios educacionais como atraso escolar e abandono.


## Comparação de Frutal com Minas Gerais e com a Região Intermediária de Uberaba

Nesta etapa, será analisado o município de Frutal em comparação com dois recortes: o conjunto dos municípios de Minas Gerais e a Região Geográfica Intermediária de Uberaba. Essa comparação permite verificar se Frutal se aproxima mais do padrão estadual ou do padrão regional.

```python
frutal = df[
    df["NO_MUNICIPIO"].str.upper() == "FRUTAL"
]

variaveis_comparacao = [
    "MEDIA_INSE",
    "TDI_ANOS_FINAIS",
    "MEDIA_ALUNOS_TURMA_ANOS_FINAIS",
    "PERC_ESCOLAS_BAIXA_REGULARIDADE_DOCENTE",
    "PERC_ESCOLAS_MEDIA_BAIXA_REGULARIDADE_DOCENTE",
    "PERC_ESCOLAS_MEDIA_ALTA_REGULARIDADE_DOCENTE",
    "PERC_ESCOLAS_ALTA_REGULARIDADE_DOCENTE",
    "TAXA_ABANDONO_ANOS_FINAIS"
]

comparacao_frutal = pd.DataFrame({
    "Indicador": variaveis_comparacao,
    "Frutal": frutal[variaveis_comparacao].iloc[0].values,
    "Média MG": df[variaveis_comparacao].mean().values,
    "Mediana MG": df[variaveis_comparacao].median().values,
    "Média RGInt Uberaba": df_rgint_uberaba[variaveis_comparacao].mean().values,
    "Mediana RGInt Uberaba": df_rgint_uberaba[variaveis_comparacao].median().values
})

comparacao_frutal["Diferença Frutal - Média MG"] = (
    comparacao_frutal["Frutal"] - comparacao_frutal["Média MG"]
)

comparacao_frutal["Diferença Frutal - Média Regional"] = (
    comparacao_frutal["Frutal"] - comparacao_frutal["Média RGInt Uberaba"]
)

comparacao_frutal
```

```python
ranking_mg = df.sort_values(
    by="TAXA_ABANDONO_ANOS_FINAIS",
    ascending=False
).reset_index(drop=True)

posicao_frutal_mg = ranking_mg[
    ranking_mg["NO_MUNICIPIO"].str.upper() == "FRUTAL"
].index[0] + 1

ranking_regiao = df_rgint_uberaba.sort_values(
    by="TAXA_ABANDONO_ANOS_FINAIS",
    ascending=False
).reset_index(drop=True)

posicao_frutal_regiao = ranking_regiao[
    ranking_regiao["NO_MUNICIPIO"].str.upper() == "FRUTAL"
].index[0] + 1

print("Posição de Frutal em MG, do maior para o menor abandono:", posicao_frutal_mg)
print("Total de municípios em MG:", len(df))

print("Posição de Frutal na Região Intermediária de Uberaba:", posicao_frutal_regiao)
print("Total de municípios na região:", len(df_rgint_uberaba))
```

A comparação entre Frutal, Minas Gerais e a Região Geográfica Intermediária de Uberaba permite contextualizar melhor a situação do município dentro do estado e de seu recorte regional.

A Média de Alunos por Turma nos Anos Finais em Frutal é de 26,2 alunos, valor superior à média estadual de 24,49 e também superior à média regional de 25,77. Embora a diferença não seja muito elevada, esse resultado indica que as turmas do município são ligeiramente mais cheias do que o padrão observado em Minas Gerais e na Região Intermediária de Uberaba.

Em relação à regularidade docente, Frutal apresenta 0,0% de escolas na faixa de baixa regularidade, 8,7% na faixa média-baixa, 87,0% na faixa média-alta e 4,3% na faixa alta. A Taxa de Abandono Escolar dos Anos Finais em Frutal é de 1,5%, valor muito próximo da média estadual de 1,47% e ligeiramente inferior à média regional de 1,62%. No ranking de abandono escolar, considerando os municípios ordenados do maior para o menor abandono, Frutal ocupa a 345ª posição entre os 853 municípios de Minas Gerais e a 13ª posição entre os 29 municípios da Região Intermediária de Uberaba. Esses resultados reforçam que Frutal se encontra em uma posição intermediária, não estando entre os municípios mais críticos, mas também não figurando entre aqueles com menor abandono.

De modo geral, Frutal apresenta um perfil educacional misto. O município possui nível socioeconômico acima da média estadual e perfil favorável de regularidade docente. Entretanto, apresenta distorção idade-série acima da média de Minas Gerais e turmas ligeiramente mais cheias. A taxa de abandono, por sua vez, está próxima da média estadual e abaixo da média regional, indicando que o município deve ser acompanhado, especialmente em relação ao atraso escolar, mas não aparece como um caso extremo de abandono.

### Interpretação da distribuição da Taxa de Abandono Escolar

Essa distribuição reforça a necessidade de transformar a taxa de abandono em classes de risco, permitindo comparar municípios com baixo, médio e alto abandono escolar.
