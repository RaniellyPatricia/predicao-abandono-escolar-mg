# Coleta e Descrição dos Dados

## Predição do Abandono Escolar nos Municípios de Minas Gerais

Este notebook tem como objetivo realizar a leitura inicial, a identificação da estrutura e a descrição das bases públicas do INEP utilizadas no projeto.

Nesta etapa ainda não serão aplicados filtros nem tratamentos definitivos. Primeiro serão verificadas as abas, colunas, dimensões, tipos de dados, valores ausentes e possíveis duplicidades.

## 1. Nível Socioeconômico — INSE

Nesta seção será realizada a inspeção inicial da base municipal do Indicador de Nível Socioeconômico de 2023.

Serão analisadas a estrutura da planilha, as abas disponíveis, as dimensões, as colunas, os tipos de dados, os valores ausentes e possíveis duplicidades. Nesta etapa, a base ainda não será filtrada ou modificada.

```python
import pandas as pd
import requests
from io import BytesIO

url_inse = (
    "https://raw.githubusercontent.com/"
    "RaniellyPatricia/predicao-abandono-escolar-mg/"
    "main/data/raw/2023/inse/INSE_2023_municipios.xlsx"
)

resposta = requests.get(url_inse)
resposta.raise_for_status()

print("Tamanho baixado:", len(resposta.content), "bytes")
print("Assinatura do arquivo:", resposta.content[:4])

arquivo_inse = pd.ExcelFile(BytesIO(resposta.content))

arquivo_inse.sheet_names
```

```python
df_inse = pd.read_excel(
    BytesIO(resposta.content),
    sheet_name="INSE_MUN_2023"
)

print("Quantidade de linhas:", df_inse.shape[0])
print("Quantidade de colunas:", df_inse.shape[1])

df_inse.head()
```

```python
print("Colunas da base INSE:")

for coluna in df_inse.columns:
    print(coluna)
df_inse.dtypes
```

```python
resumo_nulos_inse = pd.DataFrame({
    "quantidade_nulos": df_inse.isna().sum(),
    "percentual_nulos": (df_inse.isna().mean() * 100).round(2)
})

resumo_nulos_inse
```

```python
quantidade_duplicadas_inse = df_inse.duplicated().sum()
print("Quantidade de linhas duplicadas:", quantidade_duplicadas_inse)
```

```python
registros_duplicados_inse = df_inse[
    df_inse.duplicated(keep=False)
].copy()

print(
    "Total de registros envolvidos em duplicidades:",
    registros_duplicados_inse.shape[0]
)

registros_duplicados_inse.head(10)
```

```python
frequencia_duplicatas_inse = (
    df_inse
    .value_counts(dropna=False)
    .reset_index(name="FREQUENCIA")
)

frequencia_duplicatas_inse = frequencia_duplicatas_inse[
    frequencia_duplicatas_inse["FREQUENCIA"] > 1
].sort_values(
    by="FREQUENCIA",
    ascending=False
)

frequencia_duplicatas_inse.head(10)
```

### Diagnóstico das duplicidades

Foram encontradas 54.165 ocorrências duplicadas, considerando como duplicada toda linha idêntica a uma ocorrência anterior.

No total, 63.034 registros estão envolvidos em grupos de duplicidade. Após considerar apenas uma ocorrência de cada linha, a base possui 71.576 registros únicos.

Algumas linhas apresentam frequência muito elevada, chegando a 1.121 repetições de um mesmo registro.

Como todas as 19 variáveis são idênticas nesses casos, as ocorrências foram classificadas como duplicidades exatas. A remoção será realizada posteriormente, na etapa de pré-processamento, mantendo a base original inalterada.

```python
df_inse_tratado = df_inse.copy()

df_inse_tratado = df_inse_tratado[
    df_inse_tratado["SG_UF"] == "MG"
].copy()

print("Linhas da base original:", len(df_inse))
print("Linhas de Minas Gerais:", len(df_inse_tratado))

df_inse_tratado.head()
```

```python
linhas_antes = len(df_inse_tratado)

df_inse_tratado = df_inse_tratado.drop_duplicates().copy()

linhas_depois = len(df_inse_tratado)

print("Linhas antes:", linhas_antes)
print("Linhas depois:", linhas_depois)
print("Duplicidades removidas:", linhas_antes - linhas_depois)
```

```python
df_inse_tratado = df_inse_tratado[
    (df_inse_tratado["TP_TIPO_REDE"]== 6) &
    (df_inse_tratado["TP_LOCALIZACAO"] == 0)
].copy()

print("Quantidade de linhas:", len(df_inse_tratado))
print(
    "Quantidade de municípios únicos:",
    df_inse_tratado["CO_MUNICIPIO"].nunique()
)

df_inse_tratado.head()
```

```python
print(
    "Valores ausentes em MEDIA_INSE:",
    df_inse_tratado["MEDIA_INSE"].isna().sum()
)

print(
    "Percentual de valores ausentes:",
    round(df_inse_tratado["MEDIA_INSE"].isna().mean() * 100, 2),
    "%"
)
```

```python
df_inse_final = df_inse_tratado[
    [
        "NU_ANO_SAEB",
        "CO_MUNICIPIO",
        "NO_MUNICIPIO",
        "QTD_ALUNOS_INSE",
        "MEDIA_INSE"
    ]
].copy()

df_inse_final = df_inse_final.rename(
    columns={"NU_ANO_SAEB": "ANO"}
)

df_inse_final.head()
```

```python
df_inse_final.to_csv(
    "inse_mg_2023_tratado.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Arquivo salvo com sucesso.")
```
